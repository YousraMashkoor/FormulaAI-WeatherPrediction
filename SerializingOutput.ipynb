```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
```

```python
data = pd.read_csv("weatherpreprocess.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
df = data
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
```

```python
df.set_index('TIMESTAMP', inplace=True)
```

```python
df = df.sort_index()
df.head()
```

```python
# Resample and Aggregate appropriately.
x= df.resample('5T').agg({'M_WEATHER': np.median, 'M_RAIN_PERCENTAGE': np.mean})
```
